This notebook was used to create visualizations of the CNN and Yolo model results

## CNN Model (with Reddit)

```python
# Model 2 train and test accuracy scores for each epoch
accuracy = [0.5572,0.6402,0.7076,0.6877,0.7491,0.7677,0.7756,0.7972,0.8057,0.8153]
val_accuracy = [0.5704,0.7437,0.7617,0.7671,0.8051,0.7942,0.8032,0.8014,0.7888,0.8032]
```

```python
# build graph
plt.figure(figsize=(15,10))
plt.plot(accuracy, label='Training accuracy')
plt.plot(val_accuracy, label='Validation accuracy')
plt.legend();
plt.title("Accuracy Over Epochs", size=20)
plt.xlabel("Epochs", size=15)
plt.ylabel("Accuracy", size=15)
```

## YOLO Model

```python
# Yolo mAP, precision, and recall scores for each 1000 iteration
iterations = [1000,2000,3000,4000,5000,6000]
mAP = [0.840149,0.825992,0.833452,0.827290,0.815964,0.802442]
precision = [.83, .83, .86, .88, .86, .84]
recall = [.8, .8, .82, .82, .78, .78]
```

```python
# build graph
plt.figure(figsize=(15,10))
plt.plot(iterations, mAP, label='mAP')
plt.plot(iterations, precision, label='precision')
plt.plot(iterations, recall, label='recall')
plt.legend();
plt.title("Metrics By Iteration (1000)", size=20)
plt.xlabel("Iteration (1000)", size=15)
plt.ylabel("Score", size=15)
```

mAP was the highest in the first 1000 iterations followed by 3000 iterations. mAP decreased after 3000 iterations. Precision scores were higher than recall for the entire training period, showing that the model optimizes for precision.
